```python
import pandas as pd
```

```python
df = pd.read_csv("ADMISSIONS.csv")
```

```python
df.head()
```

```python
df.info()
```

```python
date_cols = ['admittime', 'dischtime', 'deathtime', 'edregtime', 'edouttime']
for col in date_cols:
    df[col] = pd.to_datetime(df[col], errors='coerce')
```

```python
df.isnull().sum()
```

```python
df = df.drop_duplicates()
```
